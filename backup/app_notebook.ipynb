```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Produção de vinhos, sucos e derivados do Rio Grande do Sul
ano = 2023
url = f"http://vitibrasil.cnpuv.embrapa.br/index.php?ano={ano}&opcao=opt_02"
response = requests.get(url)
```

```python
if response.status_code == 200:
    html_content = response.text
    print("Página obtida com sucesso")
else:
    print(f"Erro ao acessar a página. Código de status: {response.status_code}")
```

```python
# Parseia o HTML da página usando Beautifulsoup
soup = BeautifulSoup(response.text, "html.parser")

# Encontra a tabela específica pela classe
tabela = soup.find("table", {"class": "tb_base tb_dados"})

# Extrai as linhas da tabela
linhas = tabela.find_all("tr")
```

```python
# Inicializa listas para armazenar os cabeçalhos e os dados
cabecalhos = []
dados = []
```

```python
# Processa a primeira linha para extrair os cabeçalhos
# cabecalhos = [th.get_text(strip=True) for th in linhas[0].find_all("th")]

cabecalhos = ["Produto", ano]
```

```python
# Itera sobre as linhas restantes para extrair os dados
for linha in linhas[1:]:
    celulas = linha.find_all("td")
    if celulas:  # Ignora linhas sem células
        valores = [td.get_text(strip=True) for td in celulas]
        dados.append(valores)
```

```python
# Cria o DataFrame
df = pd.DataFrame(dados, columns=cabecalhos)
```
